```python
import numpy as np
import math

def Had(i):
    if i == 1:
        return(np.array([[1,1],[1,-1]]))
    else:
        return(np.kron(np.array([[1,1],[1,-1]]), Had(i-1)))

def cp(x,y):
    z=0
    for i in range(0,len(x)):
        z = (z + x[i]*y[i]) % 2
    return(z)

def cs(x,y):
    z= np.zeros((len(x),), dtype=int)
    for i in range(0,len(x)):
        z[i] = (x[i] + y[i]) % 2
    return(z)
```

```python
def alg1(k,L):
    for i in range(0,len(L)):
        if L[i] == 0:
            L[i] = 1
        else:
            L[i] = -1
    H = Had(k)
    r = np.zeros((len(L),), dtype=int)
    for i in range(0,len(L)):
        r = r + L[i]*H[i]
    return(r)
```
